```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import hddm
```

```python
# Load behavioral data
data = hddm.load_csv('HCP_wm_behavioral_data.csv')
data = data.rename(columns={'RT':'rt','accuracy':'response','subject':'subj_idx'})
data.rt = data.rt/1000
#data = data[data['response'] == 1] #Removing Error Trials
data = data[data['rt'] != 0] #Removing no-response trials
LR_data = data[data['session'] == 'LR'] #keeping only LR trials
RL_data = data[data['session'] == 'RL'] #keeping only RL trials

# Load model fits
LR_fits = hddm.load_csv('hcp_wm_HDDM_2bk_0bk_pretty_LR.csv')
RL_fits = hddm.load_csv('hcp_wm_HDDM_2bk_0bk_pretty_LR.csv')
```

```python
tasks = ['0bk' , '2bk']
def combine_data(fits,data,task1,task2):
    subs = fits['subj']
    behavior = {'mean_RT':[],'subject':[],'mean_ACC':[],'task':[],
                '%s_a'%():[],'2bk_a':[],'0bk_v':[],
                '2bk_v':[],'0bk_t':[],'2bk_t':[]}
    for task in tasks:
        for sub in subs:
            rt = data[(data['subj_idx'] == sub) & (data['task'] == task)]['rt'].mean()
            acc = data[(data['subj_idx'] == sub) & (data['task'] == task)]['response'].mean()
            behavior['0bk_a'].append(float(fits[fits['subj'] == sub]['0bk_a']))
            behavior['2bk_a'].append(float(fits[fits['subj'] == sub]['2bk_a']))
            behavior['0bk_v'].append(float(fits[fits['subj'] == sub]['0bk_v']))
            behavior['2bk_v'].append(float(fits[fits['subj'] == sub]['2bk_v']))
            behavior['0bk_t'].append(float(fits[fits['subj'] == sub]['0bk_t']))
            behavior['2bk_t'].append(float(fits[fits['subj'] == sub]['2bk_t']))
            behavior['mean_RT'].append(rt)
            behavior['mean_ACC'].append(acc)
            behavior['subject'].append(sub)
            behavior['task'].append(task)
    df = pd.DataFrame.from_dict(behavior)
    return df
LR_df = combine_data(LR_fits,LR_data,'0bk','2bk')
RL_df = combine_data(RL_fits,RL_data)
```

```python
sns.jointplot(x='0bk_v',y='mean_RT',data=LR_df,kind='reg')
sns.jointplot(x='0bk_v',y='mean_ACC',data=LR_df,kind='reg')
sns.jointplot(x='2bk_v',y='mean_RT',data=LR_df,kind='reg')
sns.jointplot(x='2bk_v',y='mean_ACC',data=LR_df,kind='reg')
```

```python
sns.jointplot(x='0bk_v',y='mean_RT',data=RL_df,kind='reg')
sns.jointplot(x='0bk_v',y='mean_ACC',data=RL_df,kind='reg')
sns.jointplot(x='2bk_v',y='mean_RT',data=RL_df,kind='reg')
sns.jointplot(x='2bk_v',y='mean_ACC',data=RL_df,kind='reg')
```
